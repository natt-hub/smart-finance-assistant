```python
import pandas as pd

# Load dataset
df = pd.read_csv('https://raw.githubusercontent.com/natt-hub/smart-finance-assistant/main/sample_transactions.csv')

# Clean Amount column
df['Amount'] = pd.to_numeric(df['Amount'].str.replace('$', ''), errors='coerce')

# Drop invalid values
df = df.dropna(subset=['Amount'])

# Basic analysis
total_spent = df['Amount'].sum()
average_spent = df['Amount'].mean()

# Group by category
category_summary = df.groupby('Category')['Amount'].sum().sort_values(ascending=False)

print("📊 SMART FINANCE SUMMARY")
print("=" * 40)
print(f"Total Spending: ${total_spent:.2f}")
print(f"Average Transaction: ${average_spent:.2f}")

print("\n💰 Spending by Category:")
for category, amount in category_summary.items():
    print(f"{category}: ${amount:.2f}")
```
